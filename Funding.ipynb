```python
# Importing libraries
import pandas as pd
file_name = 'funding.xlsx'

# Read the Excel file
df = pd.read_excel(file_name)
```

```python
df.drop(columns=['Year']).describe()
```

```python
# Summary of dataset
# Specify the columns to summarize
columns_to_summarize = ['Democratic', 'Republic']
for column in columns_to_summarize:
    df[column] = df[column].replace({'\$': '', ',': ''}, regex=True).astype(float)

summary = df[columns_to_summarize].describe()
print(summary)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Bar plot for the given data
plt.figure(figsize=(10, 6))

# Set width for the bars
bar_width = 0.35

# Set positions for the bars on the x-axis
index = df['Year']

# Plot bars for the 'Democratic' column
plt.bar(index - bar_width / 2, df['Democratic'], bar_width, label='Democratic', color='blue')

# Plot bars for the 'Republic' column
plt.bar(index + bar_width / 2, df['Republic'], bar_width, label='Republic', color='red')

# Add title and labels
plt.title('Funding Trends Over Years', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Funding', fontsize=12)

# Shorten the y-axis range
plt.ylim(bottom=0)  # Start from 0 (you can also set the top limit, if needed)
# Optionally, you can set the top limit, e.g., plt.ylim(0, 5000)

# Add legend
plt.legend()

# Show grid
plt.grid(True)
plt.legend(loc='upper right', bbox_to_anchor=(1, 1))  # Moves the legend outside the plot


# Display the plot
plt.show()

```
